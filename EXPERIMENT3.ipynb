```python
marks=(90,82,66,76,33)
max_marks=max(marks)
print(max_marks)

min_marks=min(marks)
print(min_marks)

total_numberofstudents=len(marks)
print(total_numberofstudents)

sumofmarks=sum(marks)
print(sumofmarks)

average_marks=sumofmarks/total_numberofstudents
print(average_marks)

print(sorted(marks))
```
